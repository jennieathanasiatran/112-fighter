```python
class Flower():
  def __init__(self, name, petals, price):
    '''Initialize attributes'''
    self._name = name
    self._numOfPetals = petals
    self._price = price

  def getName(self):
    '''Get name of the flower'''
    return self._name 

  def getNumOfPetals(self):
    '''Get number of petals'''
    return self._numOfPetals

  def getPrice(self):
    '''Get price of the flower'''
    return self._price

  def __str__(self):
    return f'{self._name}, price {self._price}$ and has {self._numOfPetals}'
```

```python
rose = Flower('Rose', 20, 0.5)
print(rose)
```

```python
class Person():
  def __init__(self, id, name, age, gender):
    '''Initialize attributes'''
    self._id = id
    self._name = name
    self._age = age
    self._gender = gender

  def getID(self):
    '''Get ID'''
    if self._id > 0:
        return self._id
    else:
        print("There's something wrong. Make sure that ID number is greater than 0")

  def getName(self):
    '''Get name'''
    if len(self._name) < 256:
      return self._name
    else:
      print('Number of characters are over 256')

  def getAge(self):
    '''Get age'''
    return self._age

  def getGender(self):
    '''Get gender'''
    genders = ['female', 'male']
    if self._gender.lower() in genders:
      return self._gender.capitalize()
    else:
      print('Please enter your gender again')

  def __str__(self):
    return f'Name: {self._name}\nID: {self._id}\nAge: {self._age}\nGender: {self._gender}'
```

```python
class Student(Person):
    def __init__(self, id, name, age, gender, major, studentID, GPA, scoreType):
        super().__init__(id, name, age, gender)
        self._major = major
        self._studentID = studentID
        self._GPA = GPA
        self._scoreType = scoreType

    def getMajor(self):
        return self._major.capitalize()

    def getStudentID(self):
        return self._studentID

    def getScoreType(self):
        scoreTypes = ['4 scale', '10 scale']
        if self._scoreType.lower() in scoreTypes:
          return self._scoreType.capitalize()
        else:
          return f'Please enter your score type again'

    def changeToGPA(self):
      if self.getScoreType() == '4 scale':
        return self._GPA
      else:
        if self._GPA >= 8.5 and self._GPA <= 10:
          return 4
        elif self._GPA >= 8.0 and self._GPA <= 8.4:
          return 3.5
        elif self._GPA >= 7.0 and self._GPA <= 7.9:
          return 3
        elif self._GPA >= 6.5 and self._GPA <= 6.9: 
          return 2.5
        elif self._GPA >= 5.5 and self._GPA <= 6.4:
          return 2
        elif self._GPA >= 5.0 and self._GPA <= 5.4:
          return 1.5
        elif self._GPA >= 4.5 and self._GPA <= 4.9:
          return 1
        elif self._GPA >= 0 and self._GPA <= 4.4:
          return 0
        else:
          return "Can't change to GPA"

    def getGPA(self):
        if self.changeToGPA() == 4 and self._GPA >= 9 and self._GPA <= 10:
            return 'A+'
        elif self.changeToGPA() == 4 and self._GPA >= 8.5 and self._GPA <= 8.4:
            return 'A'
        elif self.changeToGPA() == 3.5:
            return 'B+'
        elif self.changeToGPA() == 3:
            return 'B'
        elif self.changeToGPA() == 2.5: 
            return 'C+'
        elif self.changeToGPA() == 2:
            return 'C'
        elif self.changeToGPA() == 1.5:
            return 'D+'
        elif self.changeToGPA() == 1:
            return 'D'
        elif self.changeToGPA() == 0:
            return 'F'
        else:
            return "There's something wrong"

    def __eq__(self, other):
      return self._studentID == other._studentID or self._name == other._name

    def find(self, studentList):
      for student in studentList:
        if self == student:
          return self.__str__()
        else:
          return f'{self._name} is not found in the list of these students'

    def __str__(self):
          print(super().__str__())
          return f'Student ID: {self._studentID}\nMajor: {self._major}\nGPA: {self.getGPA()}'
      
```
